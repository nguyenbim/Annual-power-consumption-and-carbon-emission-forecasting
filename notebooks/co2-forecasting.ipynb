Key Features:


*   Entity: The name of the country or region for which the data is reported.
*   Year: The year for which the data is reported, ranging from 2000 to 2020.
*   Access to electricity (% of population): The percentage of population with access to electricity.
*   Access to clean fuels for cooking (% of population): The percentage of the population with primary reliance on clean fuels.
*   Renewable-electricity-generating-capacity-per-capita: Installed Renewable energy capacity per person
*   Financial flows to developing countries (US $): Aid and assistance from developed countries for clean energy projects.
*   Renewable energy share in total final energy consumption (%): Percentage of renewable energy in final energy consumption.
*   Electricity from fossil fuels (TWh): Electricity generated from fossil fuels (coal, oil, gas) in terawatt-hours.
*   Electricity from nuclear (TWh): Electricity generated from nuclear power in terawatt-hours.

*   Electricity from renewables (TWh): Electricity generated from renewable sources (hydro, solar, wind, etc.) in terawatt-hours.

*   Low-carbon electricity (% electricity): Percentage of electricity from low-carbon sources (nuclear and renewables).

*   Primary energy consumption per capita (kWh/person): Energy consumption per person in kilowatt-hours.
*   Energy intensity level of primary energy (MJ/$2011 PPP GDP): Energy use per unit of GDP at purchasing power parity.
*   Value_co2_emissions (metric tons per capita): Carbon dioxide emissions per person in metric tons.
*   Renewables (% equivalent primary energy): Equivalent primary energy that is derived from renewable sources.
*   GDP growth (annual %): Annual GDP growth rate based on constant local currency.
*   GDP per capita: Gross domestic product per person.
*   Density (P/Km2): Population density in persons per square kilometer.
*   Land Area (Km2): Total land area in square kilometers.
*   Latitude: Latitude of the country's centroid in decimal degrees.
*   Longitude: Longitude of the country's centroid in decimal degrees.

Target is Value_co2_emissions_kt_by_country for carbon emission and Primary energy consumption per capita (kWh/person) for energy consumption

Latitude, Longitude and Entity is kind of the same, since 1 country only has 1 lat and 1 long, so we can drop Lat and Long from the dataset. Density, Land Area values do not change overtime, which means that they (data of these columns) are constant, and in my opinion, they do not help to predict values change overtime like carbon emission and energy consumption (because they are likely only to help you to determine that the row is belong to which country/area and if use them as features in modelling, they can cause multicollinearity)

```python
# impute missing values by entity

def impute_missing_values_by_entity_first_step(df_fillna):
    df = df_fillna.copy()
    for entity in df['Entity'].unique():
        entity_df = df[df['Entity'] == entity]
        # fill all nan values in the beginning and final of all columns with interpolate linear, limit_direction both, limit_area outside
        entity_df = entity_df.interpolate(limit_direction='both', limit_area='outside')
        # fill all nan values in the middle of all columns with interpolate linear, limit_direction both, limit_area inside
        entity_df = entity_df.interpolate(limit_direction='both', limit_area='inside')
        df[df['Entity'] == entity] = entity_df
    df = df.reset_index(drop=True)
    return df

df_fillna_first_step = impute_missing_values_by_entity_first_step(df_fillna)
```

```python
# Hàm tạo time series cho mỗi nhóm dữ liệu
def create_time_steps(group):
    entity = group['Entity'].iloc[0]
    cols = group.columns[2:-5]
    ts = pd.DataFrame()
    # add column Entity to ts
    ts['Entity'] = [entity]
    for col in cols:
        # Add cột mới với tên là tên cột cũ + '_t-1'
        ts[col + '_t'] = group[col].values[2]
        ts[col + '_t-1'] = group[col].values[1]
        ts[col + '_t-2'] = group[col].values[0]
    ts['CO2_t+1'] = group['CO2_1'].values[0]
    ts['CO2_t+2'] = group['CO2_2'].values[0]
    ts['CO2_t+3'] = group['CO2_3'].values[0]
    ts['CO2_t+4'] = group['CO2_4'].values[0]
    ts['CO2_t+5'] = group['CO2_5'].values[0]

    return ts

train_ts = pd.DataFrame()
test_ts = pd.DataFrame()

for i in range(train_data.shape[0]//3):
    group = train_data.iloc[i*3:i*3+3]
    ts = create_time_steps(group)
    if i == 0:
        train_ts = ts
    else:
        train_ts = pd.concat([train_ts, ts])


for i in range(test_data.shape[0]//3):
    group = test_data.iloc[i*3:i*3+3]
    ts = create_time_steps(group)
    if i == 0:
        test_ts = ts
    else:
        test_ts = pd.concat([test_ts, ts])

# reset index for train_ts and test_ts
train_ts.reset_index(drop=True, inplace=True)
test_ts.reset_index(drop=True, inplace=True)
```

```python
def create_fold_ids(dataframe, n_folds):
    # Tạo bản sao của dataframe để tránh sửa đổi trực tiếp
    df_copy = dataframe.copy()
    
    # Tạo danh sách duy nhất các entity
    unique_entities = df_copy['Entity'].unique()

    # Chia fold cho mỗi entity
    for entity in unique_entities:
        entity_data = df_copy[df_copy['Entity'] == entity]
        fold_res = len(entity_data) % n_folds
        fold_size = (len(entity_data) - fold_res) // n_folds

        # Tạo fold_id cho mỗi dòng của entity
        fold_ids = np.arange(n_folds)
        np.random.shuffle(fold_ids)

        # Tạo mảng fold_id cho tất cả các dòng của entity
        all_fold_ids = np.repeat(fold_ids, fold_size)

        # Thêm các giá trị fold_id cho các dòng còn lại
        all_fold_ids = np.concatenate((all_fold_ids, np.random.choice(fold_ids, fold_res)))

        # Cập nhật fold_id trong DataFrame
        df_copy.loc[df_copy['Entity'] == entity, 'fold_id'] = all_fold_ids

    return df_copy
```

```python
import tensorflow as tf
from keras import backend as K

def custom_weighted_mse(y_true, y_pred):
    # Tính squared error cho mỗi dự đoán
    squared_error = tf.keras.backend.square(y_true - y_pred)

    # Tạo trọng số: giá trị ngược của thời gian dự đoán, nhân với một hằng số
    weights = 1 / (tf.range(1, squared_error.shape[1] + 1, dtype=tf.float32)*2)*1000

    # Nhân squared error với trọng số
    weighted_squared_error = squared_error * weights

    # Tính mean của squared error cho mỗi mốc thời gian
    loss = tf.keras.backend.mean(weighted_squared_error, axis=1)

    # Tính mean của các mean squared error cho từng mốc thời gian
    loss = tf.keras.backend.mean(loss)

    return loss
```

```python
from sklearn.metrics import mean_squared_error
from matplotlib import pyplot
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers import LSTM
from keras.layers import RepeatVector
from keras.layers import TimeDistributed
from keras.layers import Dropout
from keras.regularizers import l2
from keras.layers import LeakyReLU
from keras.callbacks import EarlyStopping

# define model


def model_build(train_features, n_lags, n_vecs=100, regularizer_alpha=0.001, dropout=0.2, n_units_1=200, n_units_2=200):
    regularizer = l2(regularizer_alpha)
    model = Sequential()
    model.add(LSTM(n_units_1,activation='relu', input_shape=(n_lags, train_features), kernel_regularizer=regularizer))
    model.add(Dropout(dropout))
    model.add(RepeatVector(5))
    model.add(LSTM(n_units_2,activation='relu', return_sequences=True, kernel_regularizer=regularizer))
    model.add(Dropout(dropout))
    model.add(TimeDistributed(Dense(n_vecs, activation='relu', kernel_regularizer=regularizer)))
    model.add(TimeDistributed(Dense(1, kernel_regularizer=regularizer)))
    model.compile(loss=custom_weighted_mse , optimizer='adam')
    return model


```
